Cread y√ßuna columna nueva y una funci√≥n que devuelva unicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.

Lo hemos resuelto de una manera alternativa porque no sab√≠amos c√≥mo hacerlo con una funcion y apply.

```python
lista_N_Y_NaN = []  

def sacar_NoY(x):                        #Con esta funcion a√±adimos a la lista los valores que nos interesan y sustituimos 
    for letra in x:                      #  los que no por NaN.
        if letra == 'N':
            lista_N_Y_NaN.append(letra)
        elif letra == 'Y':
            lista_N_Y_NaN.append(letra)
        else:
            lista_N_Y_NaN.append('NaN')
    print(lista_N_Y_NaN)
sacar_NoY(ataques_copia['Fatal (Y/N)'])
```

```python
lista_fatal_nuevos = pd.DataFrame(lista_N_Y_NaN)   #Creamos un dataframe de la lista.
lista_fatal_nuevos
```
