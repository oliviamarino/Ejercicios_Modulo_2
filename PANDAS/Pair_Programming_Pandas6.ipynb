```python
import pandas as pd
```

```python
lectura_attacks_todos = pd.read_csv("attacks_todos.csv", index_col = 0)
lectura_attacks_todos.head()
```

### EJERCICIO 1
¿Cuántos ataques hubo por país? Devuelve un dataframe.
Ordena los resultados de mayor a menor. Un dataframe lo podemos ordenar con .sort_values().

```python
df_groupby_pais = lectura_attacks_todos.groupby('Country').size().reset_index()
df_groupby_pais = df_groupby_pais.rename({0: 'num_ataques'}, axis=1)
df_groupby_pais
df_groupby_pais.sort_values(by= ['num_ataques'], ascending=False)
```

### EJERCICIO 2
¿Cuántos ataques ha habido por año? Devuelve un dataframe. De nuevo ordena los resultados de mayor a menor.

```python
df_groupby_año = lectura_attacks_todos.groupby('Year').size().reset_index()
df_groupby_año
df_groupby_año = df_groupby_año.rename({0: 'cantidad_ataques'}, axis=1)
df_groupby_año.sort_values(by= ['cantidad_ataques'], ascending=False)
```

### EJERCICIO 3
¿Cuántos ataques hubo por año y sexo? Devolved un dataframe solo del año 2000 en adelante.

```python
df_groupby_año = lectura_attacks_todos.groupby(['Year', 'Sex ']).size().reset_index()
df_groupby_año2000 = df_groupby_año[df_groupby_año['Year']>=2000]
df_groupby_año = df_groupby_año.rename({0: 'cantidad_ataques_año_sexo'}, axis=1)
df_groupby_año
```

### EJERCICIO 4
Primero extraed solo las filas que correspondan a USA, AUSTRALIA y SOUTH AFRICA. ¿Cuántos de los ataques fueron letales por sexo? Devuelve un dataframe.

```python
filtrado_paises = lectura_attacks_todos[(lectura_attacks_todos['Country'] == 'AUSTRALIA') 
                  |(lectura_attacks_todos['Country'] == 'USA')|(lectura_attacks_todos['Country'] == 'SOUTH AFRICA')]
filtrado_groupby_sexo = filtrado_paises.groupby(['Country','Sex ', 'Fatal (Y/N)']).size().reset_index()
filtrado_groupby_sexo = filtrado_groupby_sexo.rename({0: 'cantidad_ataques_letales'}, axis=1)
filtrado_groupby_sexo
```

### EJERCICIO 5
Usando el dataframe creado en el ejercicio anterior. ¿En que pais hubo más ataques?

```python
maximo_ataques = filtrado_groupby_sexo.max().reset_index()
maximo_ataques = maximo_ataques.rename({0: 'pais_max_ataques'}, axis=1)
maximo_ataques
```
