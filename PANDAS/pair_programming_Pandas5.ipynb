EJERCICIO 3. En este momento nos habremos dado cuenta que el dataframe que generamos en el pair programming Pandas III es similar al dataframe de los datos historicos. Realizad los cambios que creais necesarios para que los dataframes sean iguales y luego juntadlos usando el mejor método.

```python
#Hemos comprobado que en el csv historico tenemos columnas que no necesitamos, por lo que procedemos a borrarlas.
#De esta forma, tendremos dos csv con el mismo numero y nombre de columnas.
ataques_historico_modificado = ataques_historico.drop(["Unnamed: 22", "Unnamed: 23", "Case Number.1", "Case Number.2", "original order", "Investigator or Source",
 "pdf", "href","href formula"], axis = 1)

ataques_historico_modificado.tail()
```

EJERCICIO 4. Es el momento de unir la información de todos los ataques de tiburón que acabamos de crear con el dataframe de información adicional.

```python
csv_concatenado = pd.concat([ataques_modificado, ataques_historico_modificado], axis=0, ignore_index=True)
csv_concatenado.tail()
```

```python
#Vemos que coincide sumar las filas de los dos csv.
csv_concatenado.shape
```

```python
#Para unir el ultimo csv, como no tienen el mismo número de columnas, lo concatenaremos con merge,
#por la columna en común 'case number'.
print(csv_concatenado.columns)
print(ataques_adicional.columns)
print(ataques_adicional.shape)
```

```python
todos_csv = csv_concatenado.merge(ataques_adicional, how='inner', on='Case Number')
todos_csv.tail()
todos_csv.to_csv("attacks_todos.csv")
```
