EJERCICIO 3: ¿Cuál es el número de ataques que se han producido en España? ¿En qué regiones se han producido los ataques? ¿Cuántos ataques se han producido en cada área?

```python
lectura_attacks['Country'].unique()
numero_ataques_ES = lectura_attacks[(lectura_attacks['Country'] == 'SPAIN')]
print('El total de ataques en España han sido:', numero_ataques_ES.shape[0],'\n')

ataques_ES_region = numero_ataques_ES['Location'].unique()
print('Las regiones donde se han producido ataques en España han sido:', '\n',ataques_ES_region,'\n') 

ataques_ES_area = numero_ataques_ES['Area'].unique()
print('El número de ataques por área en España han sido:', len(ataques_ES_area)) #pongo len en lugar de shape porque sale como (5,)
```

EJERCICIO 4: Extrae un dataframe nuevo con los ataques que se han producido en Australia y Estados Unidos usando filtros e isin

```python
aus_usa_filtrado = lectura_attacks[(lectura_attacks['Country'] == 'AUSTRALIA') |(lectura_attacks['Country'] == 'USA')]
aus_usa_filtrado  #usando dos filtros
```

```python
aus_usa_isin = lectura_attacks[lectura_attacks['Country'].isin(['AUSTRALIA', 'USA'] )]
aus_usa_isin #usando isin
```

¿Ha habido ataques en Australia a gente que estuviera remando (lo encontrareis como Paddling en el dataframe en la columna activity)?

```python
ataques_aus_remo = aus_usa_isin[(aus_usa_isin['Country'] == 'AUSTRALIA') & (aus_usa_isin['Activity'] == 'Paddling')]
print('Ha habido', len(ataques_aus_remo),'ataques a personas remando en Australia')
```

¿Cuántos ataques se han producido en Australia antes del 2000?

```python
ataques_aus_2000 = aus_usa_isin[(aus_usa_isin['Country'] == 'AUSTRALIA') & (aus_usa_isin['Year'] <= 2000)]
print('Hubo', len(ataques_aus_2000),'ataques en Australia antes del año 2000')
```

¿Y después del 2000?

```python
ataques_aus_2000 = aus_usa_isin[(aus_usa_isin['Country'] == 'AUSTRALIA') & (aus_usa_isin['Year'] >= 2000)]
print('Hubo', len(ataques_aus_2000),'ataques en Australia a partir del año 2000')
```
