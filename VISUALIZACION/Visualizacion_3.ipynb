```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### EJERCICIO 1:
Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho una actividad en cada año. Hacedlo solo desde el 2015.

```python
ataques = pd.read_csv("attacks_todos.csv")
ataques.head()
```

```python
años = ataques[ataques['Year']>= 2015]
años.head()
```

```python
años.Year.unique()
```

```python
años_prueba = años.groupby(['Year','Activity']).count().reset_index()
años_prueba
```

```python
plt.figure(figsize =(8,4))
fig = sns.histplot(x=años, data = ataques,hue = 'island', bins = 20)
fig.set(xlabel='Masa corporal en g', ylabel = 'Cuentas')
fig.set_title('Histograma de los pesos de los pingüinos, en función de la isla')
plt.show()
```
